# Importing the libraries

```python
import numpy as np 
import pandas as pd
import math
```

# Importing the dataset

```python
ds = pd.read_csv('17 Batch-MasterData.csv')
```

# Checking number of null values

```python
ds.isnull().sum()
```

# Replacing null values with mean

```python
ds['10TH CGPA / %']=ds['10TH CGPA / %'].replace(np.NaN, ds['10TH CGPA / %'].mean())
```

```python
ds.isnull().sum()
```

```python
ds['INTER']=ds['INTER'].replace(np.NaN, ds['INTER'].mean())
```

```python
ds.isnull().sum()
```

# Filling categorical null values with most frequent value

```python
ds['Entrance Mode for Admission into VNR']=ds['Entrance Mode for Admission into VNR'].fillna('EAMCET')
```

```python
ds.isnull().sum()
```

# Filling all the empty semester null value with mean

```python
ds['SEM_1']=ds['SEM_1'].replace(np.NaN, ds['SEM_1'].mean())
```

```python
ds.isnull().sum()
```

```python
ds['SEM_2']=ds['SEM_2'].replace(np.NaN, ds['SEM_2'].mean())
```

```python
ds['SEM_3']=ds['SEM_3'].replace(np.NaN, ds['SEM_3'].mean())
```

```python
ds['SEM_4']=ds['SEM_4'].replace(np.NaN, ds['SEM_4'].mean())
```

```python
ds.isnull().sum()
```

# Creating new column with all the four semester average

```python
ds['SEM_4_REAL_AVG']=(ds['SEM_1']+ds['SEM_2']+ds['SEM_3']+ds['SEM_4'])/4
```

```python
ds.isnull().sum()
```

```python
ds
```

```python
print(ds['10TH BOARD NAME'].unique())
print(ds['12TH / DIPLOMA BOARD NAME'].unique())
print(ds['Entrance Mode for Admission into VNR'].unique())
print(ds['STUDENT CITY'].unique())
print(ds['Registration for Placements/ Higher Education'].unique())
```

# Cleaning the categorical column

```python
ds['STUDENT CITY'] = ds['STUDENT CITY'].str.lower()
ds['Registration for Placements/ Higher Education']=ds['Registration for Placements/ Higher Education'].str.lower()
ds['CATEGORY'] = ds['CATEGORY'].str.lower()
ds['Registration for Placements/ Higher Education'] = ds['Registration for Placements/ Higher Education'].str.replace(" ","")
ds['CATEGORY']=ds['CATEGORY'].fillna('oc')
```

```python
print(ds['Registration for Placements/ Higher Education'].unique())
print(ds['STUDENT CITY'].unique())
print(ds['CATEGORY'].unique())
```

# Converting categorical column into one hot encoding

```python
one_hot_encoded_data = pd.get_dummies(ds, columns = ['10TH BOARD NAME', '12TH / DIPLOMA BOARD NAME','Entrance Mode for Admission into VNR','Registration for Placements/ Higher Education','CATEGORY','GENDER'])
print(one_hot_encoded_data)
```

```python
one_hot_encoded_data
```

# Merging of Ranks

```python
x=0

a=one_hot_encoded_data.index[one_hot_encoded_data['Entrance Mode for Admission into VNR_EAMCET'] == 1].tolist()
print(a)
for x in a:
    print(x)
    one_hot_encoded_data['JEE RANK'][x]=np.nan

    
```

```python
a=one_hot_encoded_data.index[one_hot_encoded_data['Entrance Mode for Admission into VNR_JEE'] == 1].tolist()
print(a)
for x in a:
    print(x)
    one_hot_encoded_data['EAMCET RANK'][x]=np.nan
```

```python
b=one_hot_encoded_data.index[one_hot_encoded_data['Entrance Mode for Admission into VNR_NRI'] == 1].tolist()
print(b)
for y in b:
    print(y)
    if one_hot_encoded_data['Entrance Mode for Admission into VNR_EAMCET'][y] == 1 and one_hot_encoded_data['Entrance Mode for Admission into VNR_JEE'][y]==1:
           one_hot_encoded_data['JEE RANK'][x]=np.nan
    
    
```

```python
e=one_hot_encoded_data.index[one_hot_encoded_data['Entrance Mode for Admission into VNR_NRI'] == 1].tolist()
print(e)
for y in e:
    print(y)
    if one_hot_encoded_data['Entrance Mode for Admission into VNR_EAMCET'][y] == 0 and one_hot_encoded_data['Entrance Mode for Admission into VNR_JEE'][y]==0 :
           one_hot_encoded_data['JEE RANK'][y]=np.nan

```
